```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import cross_val_score,KFold
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=1)
scores = cross_val_score(knn, X, y, cv=kfold, scoring='r2')
```

```python
scores.mean()
```

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsRegressor()
```

```python
param_grid = {
    'n_neighbors':[1,3,5,7,10,12,15,17,20],
    'weights':['uniform','distance'],
    'algorithm':['ball_tree', 'kd_tree', 'brute'],
    'p':[1,2]
}
```

```python
gcv = GridSearchCV(knn, param_grid, scoring='r2', refit=True, cv=kfold, verbose=2)  # verbose = 2 means that it will print the progress of the grid search. 
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
gcv.cv_results_
```

```python
pd.DataFrame(gcv.cv_results_)[['param_algorithm',	'param_n_neighbors',	'param_p', 'param_weights', 'mean_test_score']].sort_values('mean_test_score',ascending=False)
```

It is important to observe the above dataframe because in some cases , the best parameters are computationally expensive so we can choose the second best parameters which are computationally less expensive and gives us almost the same performance as the best parameters.

for example, if the neighbors = 5 gives us an accuracy of 0.95 and neighbors = 3 gives us an accuracy of 0.94 then we can choose neighbors = 3 because it is computationally less expensive than neighbors = 5 and gives us almost the same performance as neighbors = 5.
